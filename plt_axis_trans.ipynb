```python
import matplotlib.pyplot as plt
import numpy
m=[i*0.05 for i in range(0,21)]
n=[i*0.05 for i in range(10,21)]
y=[i*0.01 for i in range(0,len(m)*len(n))]

dots=[]
for i in m:
    for j in n:
        dots.append([i,j])

dots=numpy.array(dots)
print(dots.size)
p1=plt.figure(1)
# print(dots)

for i in range(0,(int)(dots.size/2)):
    plt.scatter(dots[i][0],dots[i][1],label=y[i])

# plt.scatter(dots[:,0],dots[:,1],y)
# plt.legend(loc='best')
plt.show()


for i in dots:

    xx=i[0]
    yy=i[1]
    new_y=xx+yy-xx*yy
    new_x=2*(1-new_y)*i[0]/(1-i[0])*new_y
    new_y=yy
    i[0]=new_x
    i[1]=new_y

p2=plt.figure(2)
print(y)
for i in range(0,(int)(dots.size/2)):
    plt.scatter(dots[i][0],dots[i][1],label=y[i])
# plt.scatter(dots[:,0],dots[:,1],y)
plt.show()
```
